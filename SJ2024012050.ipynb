```python
import tkinter as tk
from owlready2 import get_ontology, sync_reasoner_pellet

# Load Ontology
ontology_file = "example.owl"
ontology = get_ontology(ontology_file).load()

# Synchronize reasoner (to ensure ontology logic is updated)
with sync_reasoner_pellet():
    pass

# Get Classes from Ontology
shapes = [cls.name for cls in ontology.classes()]

# Function to Calculate Area
def calculate_area():
    shape = shape_var.get()
    try:
        if shape == "Triangle":
            base = float(base_entry.get())
            height = float(height_entry.get())
            area = 0.5 * base * height
            result_label.config(text=f"Area of {shape}: {area}")
        elif shape == "Rectangle":
            length = float(length_entry.get())
            width = float(width_entry.get())
            area = length * width
            result_label.config(text=f"Area of {shape}: {area}")
        else:
            result_label.config(text="Shape not supported.")
    except ValueError:
        result_label.config(text="Invalid input. Please enter numbers.")

# GUI Application
app = tk.Tk()
app.title("Intelligent Tutoring System")

# Shape Selection
tk.Label(app, text="Select Shape:").grid(row=0, column=0, padx=10, pady=10)
shape_var = tk.StringVar(value=shapes[0])
shape_menu = tk.OptionMenu(app, shape_var, *shapes)
shape_menu.grid(row=0, column=1, padx=10, pady=10)

# Input Fields
tk.Label(app, text="Base:").grid(row=1, column=0, padx=10, pady=5)
base_entry = tk.Entry(app)
base_entry.grid(row=1, column=1, padx=10, pady=5)

tk.Label(app, text="Height:").grid(row=2, column=0, padx=10, pady=5)
height_entry = tk.Entry(app)
height_entry.grid(row=2, column=1, padx=10, pady=5)

tk.Label(app, text="Length:").grid(row=3, column=0, padx=10, pady=5)
length_entry = tk.Entry(app)
length_entry.grid(row=3, column=1, padx=10, pady=5)

tk.Label(app, text="Width:").grid(row=4, column=0, padx=10, pady=5)
width_entry = tk.Entry(app)
width_entry.grid(row=4, column=1, padx=10, pady=5)

# Calculate Button
calculate_btn = tk.Button(app, text="Calculate Area", command=calculate_area)
calculate_btn.grid(row=5, column=0, columnspan=2, pady=10)

# Result Label
result_label = tk.Label(app, text="Result will appear here.")
result_label.grid(row=6, column=0, columnspan=2, pady=10)

# Run Application
app.mainloop()
```
